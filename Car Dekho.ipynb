```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('car data.csv')
```

```python
df.head()
```

```python
print(df['Seller_Type'].unique())
print(df['Transmission'].unique())
print(df['Owner'].unique())
```

```python
## Checking missing or null values

df.isnull().sum()
```

```python
# Creating an other column, no of years old
df['Years Old'] = 2020 - df['Year']
```

```python
df['Years Old']
```

```python
final_dataset = df[['Years Old','Selling_Price','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']]
```

```python
final_dataset.head()
```

```python
final_dataset = pd.get_dummies(final_dataset,drop_first = True)
```

```python
final_dataset.head()
```

```python
final_dataset.corr()
```

```python
import seaborn as sns
```

```python
sns.pairplot(final_dataset)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
corrmat = final_dataset.corr()
top_corr_features = corrmat.index
plt.figure(figsize = (20,20))
# plot heat map

g = sns.heatmap(final_dataset[top_corr_features].corr(), annot = True, cmap = 'RdYlGn')
```

```python
final_dataset.head()
```

```python
final_dataset['No Years'] = final_dataset['Years Old']
final_dataset.drop('Years Old', axis = 1, inplace = True)
```

```python
final_dataset.head()
```

```python
# Independent and dependent features
x = final_dataset.iloc[:,1:]
y = final_dataset.iloc[:,0]
```

```python
# Feature Importance

from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
# plot graph of feature importances for better visualization

feat_importances = pd.Series(model.feature_importances_,index = x.columns)
feat_importances.nlargest(5).plot(kind = 'barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Randomized Search CV

# Number of Trees in random Forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]

# Number of featurs to consider at every split
max_features = ['auto', 'sqrt']

# Maximum number f levels in a tree
max_depth = [int(x) for x in np.linspace(5,30,6)]

# Minimum number of samples required to split a node
min_samples_split = [2,5,10,15,100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1,3,5,10]
```

```python
# Create the random grid

random_grid = {
                'n_estimators': n_estimators,
                'max_features': max_features,
                'max_depth' : max_depth,
                'min_samples_split' : min_samples_split,
                'min_samples_leaf' : min_samples_leaf
              }
print(random_grid)
```

```python
# Use random grid to search for best hyperparameters
# First Create Base model to tune

rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, scoring = 'neg_mean_squared_error', n_iter = 10, cv = 5, verbose = 2, random_state = 42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions=rf_random.predict(X_test)
```

```python
sns.distplot(y_test-predictions)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error : ',metrics.mean_absolute_error(y_test,predictions))
print('Mean Squared Error : ',metrics.mean_squared_error(y_test,predictions))
print('Root Mean Squared Error : ',np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
import pickle
# open a file, where you ant to store the data
file = open('random_forest_regression_model.pkl', 'wb')

# dump information to that file
pickle.dump(rf_random, file)
```
